```python
import os
import random

def get_random_image_path(folder_path):
  """Gets the path to a random image in a specific folder.

  Args:
    folder_path: The path to the folder containing the images.

  Returns:
    The path to a random image in the folder.
  """

  image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
  random_image_file = random.choice(image_files)
  random_image_path = os.path.join(folder_path, random_image_file)
  return random_image_path
```

```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2

# Load the trained YOLO model
model = YOLO("runs/segment/train/weights/best.pt")

# Load a random image
folder_path = "/content/contrails-detection-2/test/images"
image_path = get_random_image_path(folder_path)
image = Image.open(image_path)

# Perform object detection
results = model(image_path)

# Plot and display results
for result in results:
    result.save("output.jpg")  # Save output if needed
    img_with_masks = result.plot()  # Get image with detections
    img_with_masks_rgb = cv2.cvtColor(img_with_masks, cv2.COLOR_BGR2RGB) # Convert the image to RGB

    # Display original image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
    plt.imshow(image)
    plt.title("Original Image")
    plt.axis("off")

    # Display image with labels
    plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
    plt.imshow(img_with_masks_rgb, cmap='jet')
    plt.title("Image with Labels")
    plt.axis("off")

    plt.show()
```
